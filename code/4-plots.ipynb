```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


%matplotlib inline
```

# jokes v dadjokes

```python
# score on TEST data

baseline = 0.508547
lr = 0.6275197195442594
nb = 0.6148115687992989
svm = 0.7598597721297108

# Set size of figure: eight by five.
fig, ax = plt.subplots(figsize=(7, 5))

plt.yticks(np.arange(0, 1.1, step=0.1))

plt.ylim(0,1.1)
# Create a bar chart.
# plt.bar() will create a vertical bar chart.
ax.bar(['Baseline','Log Regression', 'Multi Naive Bayes', 'Linear SVM'], #X labels
        [baseline, lr, nb, svm]) # Y heights of each bar


# Make some labels.
labels = [.51, .63, .61, .76]
rects = ax.patches

for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 5, label,
            ha='center', va='bottom')

# freq_series = pd.Series([baseline, lr, nb, svm])
# ax = freq_series.plot(kind='bar')


# Add a title and show the plot.
plt.title('Testing Data Score of Each Model');

plt.savefig("../plots/dadjokes")
```
